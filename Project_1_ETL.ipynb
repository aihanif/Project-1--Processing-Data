```python
def detect_browser(user_agent):
    user_agent = str(user_agent).lower()
    if 'chrome' in user_agent and 'edg' not in user_agent:
        return 'Chrome'
    elif 'firefox' in user_agent:
        return 'Firefox'
    elif 'safari' in user_agent and 'chrome' not in user_agent:
        return 'Safari'
    elif 'edg' in user_agent:
        return 'Edge'
    elif 'opera' in user_agent or 'opr' in user_agent:
        return 'Opera'
    else:
        return 'Unknown'
    

# Apply the function to the 'BROWSER_NAME' column
empty_browser_rows = datatable[datatable['BROWSER_NAME'].astype(str).str.strip().eq('')]
print(empty_browser_rows.shape) # check empty browser rows shape

```

```python
#fill data for BROWSER_NAME from BROWSER_NAME_VERSION not empty
datatable['BROWSER_NAME'] = datatable['BROWSER_NAME_VERSION'].apply(detect_browser)
#check empty browser name  filling the data from filling BROWSER_NAME_VERSION which is not empty
mask = (
    datatable['BROWSER_NAME'].fillna('').astype(str).str.strip() == ''
) & (
    datatable['BROWSER_NAME_VERSION'].notnull()
)

# fill BROWSER_NAME based on function to detect_browser from BROWSER_NAME_VERSION data
datatable.loc[mask, 'BROWSER_NAME'] = datatable.loc[mask, 'BROWSER_NAME_VERSION'].apply(detect_browser)


empty_browser_rows = datatable[datatable['BROWSER_NAME'].astype(str).str.strip().eq('')]
print(empty_browser_rows.shape) # check empty browser rows shape

print(datatable['BROWSER_NAME'].isnull().sum()) # check null value in BROWSER_NAME column
print(datatable['BROWSER_NAME'].astype(str).str.strip().eq('').sum()) # check empty value in BROWSER_NAME column
datatable
```

```python
#function to set AM/PM to data CREATED_DATE if the value is not in AM/PM format
def add_ampm(dt_str):
    try:
        #  parse as 24 hour
        dt = pd.to_datetime(dt_str, format='%d/%m/%Y %H:%M')
        # add AM/PM following 12 hour format
        ampm = 'AM' if dt.hour < 12 else 'PM'
        return dt.strftime('%d/%m/%Y %I:%M ') + ampm
    except:
        return dt_str  # back to original if error
#save in another variable    
without_ampm = no_ampm_rows.copy()
without_ampm['CREATED_DATE'] = without_ampm['CREATED_DATE'].apply(add_ampm)
without_ampm
```

```python
#check data CREATED_DATE format
Data_cleaned_2 = without_ampm
has_ampm = Data_cleaned_2['CREATED_DATE'].astype(str).str.upper().str.contains('AM|PM')
no_ampm_rows = Data_cleaned_2[~has_ampm]
print(no_ampm_rows.shape)
```

```python
#function to convert CREATED_DATE to datetime format
def flexible_datetime_parser(x):
    try:
        return pd.to_datetime(x, format='%d/%m/%Y %I:%M %p')
    except:
        try:
            return pd.to_datetime(x, format='%d/%m/%Y %H:%M')
        except:
            return pd.NaT
        
#convert CREATED_DATE to datetime format
#apply function to the column
Data_cleaned_2['CREATED_DATE'] = Data_cleaned_2['CREATED_DATE'].apply(flexible_datetime_parser)
```
